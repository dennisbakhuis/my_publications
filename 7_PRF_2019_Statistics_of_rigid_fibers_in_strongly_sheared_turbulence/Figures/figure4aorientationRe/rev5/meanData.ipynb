```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inlinex
```

```python
data = [pd.read_csv(x) for x in glob.glob('*.csv')]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
for df in data:
    ax.plot(df.bins, df['hist'], '.')
ax.plot(bins, mean, 'k-')
ax.plot(bins, mean-std, 'k--')
ax.plot(bins, mean+std, 'k--')
```

```python
bins = data[0].bins.values
mean = np.vstack([x['hist'] for x in data]).mean(axis=0)
std = np.vstack([x['hist'] for x in data]).std(axis=0)
dx = bins[1]-bins[0]
bins = np.hstack([bins[0]-dx, bins, bins[-1]+dx])
mean = np.hstack([mean[-1], mean, mean[0]])
std = np.hstack([std[-1], std, std[0]])
stdmin = mean - std
stdmax = mean + std
```

```python
df = pd.DataFrame({
    'bins':bins,
    'hist':mean * 180/np.pi,
    'std':std * 180/np.pi,
    'stdmin':stdmin * 180/np.pi,
    'stdmax':stdmax * 180/np.pi
})
```

```python
df.to_csv('meanstddata.csv', index=False)
```
